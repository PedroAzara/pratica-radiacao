```python
import numpy as np

# Arrays dos dados
V_DCT = np.array([5.310, 3.179, 2.036, 1.371, 0.965, 0.769, 0.614, 0.508, 0.428, 0.365])/1000  # Em V
distancia = np.array([15, 20, 25, 30, 35, 40, 45, 50, 55, 60])/10  # Em m

# Exibir os arrays
print("V_DCT (mV):", V_DCT)
print("Distância (cm):", distancia)

```

```python
P = V_DCT/0.14  # Em W
radiancia = P/(4*np.pi*(15/1000/2)**2)  # Em W/m² 
print("Potência (W):", P)
print("Radiancia (W/m²):", radiancia)
```

```python
import matplotlib.pyplot as plt

plt.plot(distancia, radiancia, marker='o')
plt.xlabel('Distância (m)')
plt.ylabel('Radiância (W/m²)')
plt.title('Radiância em função da Distância')
plt.grid(True)
plt.show()
```

```python
# Calcular o inverso do quadrado da distância
inverso_quadrado_distancia = 1 / (distancia ** 2)

# Plotar o gráfico
plt.plot(inverso_quadrado_distancia, radiancia, marker='o')
plt.xlabel('1 / Distância² (1/m²)')
plt.ylabel('Radiância (W/m²)')
plt.title('Radiância em função do Inverso do Quadrado da Distância')
plt.grid(True)
plt.show()
```

```python
# Lista de valores de V_DCT (mV)
V_DCT_2 = np.array([
    1.326, 1.542, 1.724, 1.977, 2.260, 2.520, 
    2.769, 3.087, 3.500, 3.842, 4.210, 4.551, 
    4.963, 5.310
])/1000

# Exemplo de exibição dos valores
print("Valores de V_DCT (mV):", V_DCT_2)
```

```python
P_2 = V_DCT_2/0.14  # Em W
radiancia_2 = P_2/(4*np.pi*(15/1000/2)**2)  # Em W/m² 
print("Potência (W):", P_2)
print("Radiancia (W/m²):", radiancia_2)
```

```python
# Arrays de valores de Voltagem (V) e Corrente (mA)
voltagem = np.array([
    4.996, 5.489, 5.997, 6.480, 7.050, 7.550, 
    8.000, 8.490, 9.050, 9.500, 10.090, 10.500, 
    11.010, 11.270
])

corrente = np.array([
    1.233, 1.305, 1.360, 1.411, 1.473, 1.517, 
    1.555, 1.588, 1.656, 1.707, 1.757, 1.785, 
    1.845, 1.890
])

# Exemplo de exibição dos valores
print("Array de Voltagem (V):", voltagem)
print("Array de Corrente (A):", corrente)
```

```python
resistencia = voltagem / corrente  # Em Ohms
resistencia_corrigida = resistencia/0.5   
print("Resistência corrigida:", resistencia_corrigida)
```

```python
from sklearn.linear_model import LinearRegression

# Array de R/R300K
R_R300K = np.array([
    1.0, 1.43, 1.87, 2.24, 2.85, 3.36, 3.88,
    4.41, 4.95, 5.48, 6.03, 6.58, 7.14, 7.71,
    8.28, 8.86, 9.44, 10.03, 10.63, 11.24, 11.84,
    12.46, 13.08, 13.72, 14.34, 14.99, 15.63
])

# Array de T(K)
T = np.array([
    300, 400, 500, 600, 700, 800, 900,
    1000, 1100, 1200, 1300, 1400, 1500, 1600,
    1700, 1800, 1900, 2000, 2100, 2200, 2300,
    2400, 2500, 2600, 2700, 2800, 2900
])

# Exemplo de exibição
print("Array de R/R300K:", R_R300K)
print("Array de T (K):", T)
# Reshape T for sklearn
T_reshaped = T.reshape(-1, 1)

# Create and fit the model
model = LinearRegression()
model.fit(T_reshaped, R_R300K)

# Get the coefficients
slope = model.coef_[0]
# Plot the data and the regression line
# Reshape R_R300K for sklearn
R_R300K_reshaped = R_R300K.reshape(-1, 1)

# Create and fit the model for T(R)
model_T_R = LinearRegression()
model_T_R.fit(R_R300K_reshaped, T)

# Get the coefficients
slope_T_R = model_T_R.coef_[0]
intercept_T_R = model_T_R.intercept_

# Plot the data and the regression line
plt.scatter(R_R300K, T, color='blue', label='Data')
plt.plot(R_R300K, model_T_R.predict(R_R300K_reshaped), color='red', label='Regression Line')
plt.xlabel('R/R300K')
plt.ylabel('Temperature (K)')
plt.title('Linear Regression of Temperature vs R/R300K')
plt.legend()
plt.grid(True)
plt.show()

# Print the equation
print(f"Equation: T = {slope_T_R} * R/R300K + {intercept_T_R}")
```

```python
#Agora com a equação em mãos, eu posso aplica-la  T = 176 *R/R300K + 208.5
temperatura = 176 *resistencia_corrigida + 208.5
```

```python
import matplotlib.pyplot as plt

plt.scatter(temperatura**4, radiancia_2)
plt.xlabel('Temperatura Quártica (K⁴)')
plt.ylabel('Radiância (W/m²)')
plt.title('Radiância em função da Temperatura Quártica')
# Elevar a temperatura à quarta potência
temperatura_quartica = temperatura**4
# Reshape temperatura**4 for sklearn
temperatura_quartica = temperatura_quartica.reshape(-1, 1)
# Create and fit the model
model_temp_radiancia = LinearRegression()
model_temp_radiancia.fit(temperatura_quartica, radiancia_2)

# Get the coefficients
slope_temp_radiancia = model_temp_radiancia.coef_[0]
intercept_temp_radiancia = model_temp_radiancia.intercept_

# Plot the data and the regression line
plt.plot(temperatura**4, model_temp_radiancia.predict(temperatura_quartica), color='red', label='Regression Line')
plt.legend()

# Print the equation
print(f"Equation: Radiância = {slope_temp_radiancia} * Temperatura^4 + {intercept_temp_radiancia}")
plt.grid(True)
plt.show()
```
